```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import time
import random
import os

# Create an instance of the Edge driver outside the loop
edge_driver_directory = r'C:\Users\Jerico\Downloads\edgedriver_win64'
os.environ['PATH'] += ';' + edge_driver_directory
driver = webdriver.Edge()

def perform_google_search(query):
    try:
        # Open Google in the existing browser
        driver.get('https://www.bing.com')

        # Find the search box using its name attribute value
        search_box = driver.find_element('name', 'q')

        # Type the search query
        search_box.send_keys(query)

        # Press Enter to perform the search
        search_box.send_keys(Keys.RETURN)

        # Wait for a few seconds to let the results load (adjust as needed)
        time.sleep(5)

        # You can perform further actions here, like scraping the search results or clicking on a link.

    except Exception as e:
        print(f"An error occurred: {e}")

# Example usage with a limit of 5 searches
search_limit = 150

for _ in range(search_limit):
    # Generate a random search query (you can modify this based on your requirements)
    search_query = f"Random Searching2 {random.randint(1, 1000)}"
    
    print(f"Performing search: {search_query}")
    perform_google_search(search_query)

    # Pause for a short duration before the next search
    time.sleep(random.uniform(1, 3))

# Close the browser window when all searches are done
driver.quit()
```
